```python
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
```

```python
def func(x):
  return (np.exp(x)/x)+(2/(x+1))
h = 0.01; min_q = 1; max_q = 6
x = np.arange(min_q, max_q, h)
def diff_nogtede(f, x, h):
    derivative = (f(x + h) - f(x))/h
    return derivative

diff_nogtede(func, x, h)
plt.plot(x,diff_nogtede(func, x, h))
diff_nogtede(func, 2.6, h)
plt.plot(x, func(x))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, diff_nogtede(func, x, h))
plt.xlabel('x')
plt.ylabel('toreme')
```
